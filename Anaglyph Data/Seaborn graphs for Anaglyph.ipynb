```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_excel("data.xlsx")
```

```python
data = data.rename(columns={'numberOfPixels': 'Liczba pikseli'})
```

```python

data_dubois_asm = data[data["kindOfAlgorithm"]=="duboisAsm"]
data_dubois_python = data[data["kindOfAlgorithm"]=="duboisPython"]
data_photoshop_asm = data[data["kindOfAlgorithm"]=="photoshopAsm"]
data_photoshop_python = data[data["kindOfAlgorithm"]=="photoshopPython"]
```

```python

data_dubois_asm
```

```python
plt.figure(figsize=(16, 12))
plot = sns.lineplot(x="numberOfThreads", y="time",hue="Liczba pikseli", data=data_dubois_asm)
plot.legend(prop={'size': 15})
plot.set_xlabel('Ilo wtkw',fontsize=15);
plot.set_ylabel('Czas [s]',fontsize=15);
plot.set_title("Algorytm Dubois ASM",fontsize=20)
```

```python
plt.figure(figsize=(16, 12))
plot = sns.lineplot(x="numberOfThreads", y="time",hue="Liczba pikseli", data=data_dubois_python)
plot.legend(prop={'size': 15})
plot.set_xlabel('Ilo wtkw',fontsize=15);
plot.set_ylabel('Czas [s]',fontsize=15);
plot.set_title("Algorytm Dubois PYTHON",fontsize=20)
```

```python
plt.figure(figsize=(16, 12))
plot = sns.lineplot(x="numberOfThreads", y="time",hue="Liczba pikseli", data=data_photoshop_asm)
plot.legend(prop={'size': 15})
plot.set_xlabel('Ilo wtkw',fontsize=15);
plot.set_ylabel('Czas [s]',fontsize=15);
plot.set_title("Algorytm Photoshop ASM",fontsize=20)
```

```python
plt.figure(figsize=(16, 12))
plot = sns.lineplot(x="numberOfThreads", y="time",hue="Liczba pikseli", data=data_photoshop_python)
plot.legend(prop={'size': 15})
plot.set_xlabel('Ilo wtkw',fontsize=15);
plot.set_ylabel('Czas [s]',fontsize=15);
plot.set_title("Algorytm Photoshop PYTHON",fontsize=20)
```
